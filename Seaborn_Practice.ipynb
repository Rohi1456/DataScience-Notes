```
data2 = pd.read_csv('train2.csv')
```

```
data2.head()
```

```
sns.catplot(y='avg_training_score',x='education',data= data2,hue = 'gender',kind = 'swarm')
```

Since we can see that the plot is scattered, so to handle that, we can set the jitter to false. Jitter is the deviation from the true value. So, well set the jitter to false by using another parameter.

```
sns.catplot(y='avg_training_score',x='education',data= data2,jitter=False)
```

n the above plots, we can see that the points are overlapping each other, to eliminate this situation, we can set kind = swarm, swarm uses an algorithm that prevents the points from overlapping and adjusts the points along the categorical axis. Lets see how it looks like-
